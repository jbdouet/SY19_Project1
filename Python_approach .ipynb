```python
import pandas as pd
import tpot 
import numpy as np 
from sklearn.model_selection import train_test_split
```

## Classification 

```python
data =pd.read_csv("data/tp3_clas_app.txt", sep=" ")
```

```python
data
```

```python
train, test = train_test_split(data, test_size=0.2)
```

```python
X_train = train.drop('y', axis=1)
X_test = test.drop('y', axis=1) 
```

```python
y_train = train['y']
y_test = test['y']
```

```python
from tpot import TPOTClassifier
```

```python
tpot = TPOTClassifier(generations=20, population_size=40, verbosity=2)
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))
#tpot.export('tpot_mnist_pipeline.py')
```

## Regression 

```python
data_reg =pd.read_csv("data/tp3_reg_app.txt", sep=" ")
```

```python
train_reg, test_reg = train_test_split(data_reg, test_size=0.25)
```

```python
X_reg_train = train_reg.drop('y', axis=1)
X_reg_test = test_reg.drop('y', axis=1) 
```

```python
y_reg_train = train_reg['y']
y_reg_test = test_reg['y']
```

```python
from tpot import TPOTRegressor 
```

```python
tpot = TPOTRegressor(generations=15, population_size=40, verbosity=2)
tpot.fit(X_reg_train, y_reg_train)
print(tpot.score(X_reg_test, y_reg_test))
```
